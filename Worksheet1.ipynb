```python
#Task7
try:
    concat_x_v = np.concatenate((x, v))
    print(f"Result: {concat_x_v}")
except ValueError as e:
    print(f"ERROR: {e}")
    print("Explanation: x has shape (2,2) and v has shape (2,).")
    print("They have incompatible shapes for direct concatenation.")
    print("We need to reshape v to match dimensions.")
    # Solution: reshape v
    v_reshaped = v.reshape(1, 2)
    concat_x_v = np.vstack((x, v_reshaped))
    print(f"After reshaping v to (1,2), concatenation works:\n{concat_x_v}")
```

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])
```

```python
# PROBLEM 4: MATRIX OPERATIONS
#Task 1
A_inv = np.linalg.inv(A)
identity_result = np.dot(A, A_inv)
print(f"A^-1 = \n{A_inv}")
print(f"A · A^-1 = \n{identity_result}")
print(f"Is it identity? {np.allclose(identity_result, np.eye(2))}")
```

```python
#Task2
AB = np.dot(A, B)
BA = np.dot(B, A)
print(f"AB = \n{AB}")
print(f"BA = \n{BA}")
print(f"Are they equal? {np.array_equal(AB, BA)}")
print("Conclusion: Matrix multiplication is NOT commutative!")
```

```python
#Task4
import numpy as np
print("System:")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")
print("3x + y - z = 9")

# Coefficient matrix A
A_eq = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])

# Constants vector B
B_eq = np.array([-1, -3, 9])

print(f"\nCoefficient matrix A:\n{A_eq}")
print(f"Constants vector B: {B_eq}")

# Method 1: Using inverse
print("\n--- Method 1: Using inverse (A^-1 · B) ---")
A_eq_inv = np.linalg.inv(A_eq)
solution_inv = np.dot(A_eq_inv, B_eq)
print(f"Solution using inverse: x = {solution_inv[0]:.2f}, y = {solution_inv[1]:.2f}, z = {solution_inv[2]:.2f}")

# Method 2: Using np.linalg.solve
print("\n--- Method 2: Using np.linalg.solve ---")
solution_solve = np.linalg.solve(A_eq, B_eq)
print(f"Solution using solve: x = {solution_solve[0]:.2f}, y = {solution_solve[1]:.2f}, z = {solution_solve[2]:.2f}")

# Verification
print("\n--- Verification ---")
verification = np.dot(A_eq, solution_solve)
print(f"A · X = {verification}")
print(f"B = {B_eq}")
print(f"Correct? {np.allclose(verification, B_eq)}")
```

```python
import time
import numpy as np
size = 1_000_000
matrix_size = 1000

print(f"\nArray size: {size:,}")
print(f"Matrix size: {matrix_size}x{matrix_size}")

# 1. ELEMENT-WISE ADDITION
# Python lists
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
python_time = time.time() - start_time
print(f"Time taken: {python_time:.6f} seconds")

# NumPy arrays
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()
result_numpy = arr1 + arr2
numpy_time = time.time() - start_time
print(f"Time taken: {numpy_time:.6f} seconds")

speedup = python_time / numpy_time
print(f"\n✓ NumPy is {speedup:.2f}x FASTER than Python lists!")

# 2. ELEMENT-WISE MULTIPLICATION
# Python lists
start_time = time.time()
result_list = [list1[i] * list2[i] for i in range(size)]
python_time = time.time() - start_time
print(f"Time taken: {python_time:.6f} seconds")

# NumPy arrays
start_time = time.time()
result_numpy = arr1 * arr2
numpy_time = time.time() - start_time
print(f"Time taken: {numpy_time:.6f} seconds")

speedup = python_time / numpy_time
print(f"\n✓ NumPy is {speedup:.2f}x FASTER than Python lists!")

# 3. DOT PRODUCT
# Python lists
start_time = time.time()
result_list = sum([list1[i] * list2[i] for i in range(size)])
python_time = time.time() - start_time
print(f"Time taken: {python_time:.6f} seconds")

# NumPy arrays
start_time = time.time()
result_numpy = np.dot(arr1, arr2)
numpy_time = time.time() - start_time
print(f"Time taken: {numpy_time:.6f} seconds")

speedup = python_time / numpy_time
print(f"\n✓ NumPy is {speedup:.2f}x FASTER than Python lists!")


# 4. MATRIX MULTIPLICATION
# Python lists
matrix1 = [[i+j for j in range(matrix_size)] for i in range(matrix_size)]
matrix2 = [[i+j for j in range(matrix_size)] for i in range(matrix_size)]

start_time = time.time()
# Matrix multiplication using nested loops
result_matrix = [[0 for _ in range(matrix_size)] for _ in range(matrix_size)]
for i in range(matrix_size):
    for j in range(matrix_size):
        for k in range(matrix_size):
            result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]
python_time = time.time() - start_time
print(f"Time taken: {python_time:.6f} seconds")

# NumPy arrays
np_matrix1 = np.arange(matrix_size * matrix_size).reshape(matrix_size, matrix_size)
np_matrix2 = np.arange(matrix_size * matrix_size).reshape(matrix_size, matrix_size)

start_time = time.time()
result_np = np.dot(np_matrix1, np_matrix2)
numpy_time = time.time() - start_time
print(f"Time taken: {numpy_time:.6f} seconds")

speedup = python_time / numpy_time
print(f"\n✓ NumPy is {speedup:.2f}x FASTER than Python lists!")
```
