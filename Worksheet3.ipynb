```python
titanic = pd.read_csv("/content/drive/MyDrive/Concept and Technologies of AI/Titanic-Dataset.csv")
titanic.head()
```

```python
# 1. Fare column only
fare = titanic[['Fare']]
fare.head()
```

```python
# 2. Class + Age columns
class_age = titanic[['Pclass','Age']]
class_age.head()
```

```python
# 3. Survived + Sex columns
survived_gender = titanic[['Survived','Sex']]
survived_gender.head()
```

```python
# 1. Fare > 100
fare_gt_100 = titanic[titanic['Fare'] > 100]
fare_gt_100.head()
```

```python
# 2. First Class (Pclass = 1)
first_class = titanic[titanic['Pclass'] == 1]
first_class.head()
```

```python
# 3. Female under 18
female_under_18 = titanic[(titanic['Age'] < 18) & (titanic['Sex'] == 'female')]
female_under_18.head()
```

```python
# 1. Embarked C or S
embarked_c_or_s = titanic[titanic['Embarked'].isin(['C','S'])]
embarked_c_or_s.head()
```

```python
# 2. Pclass in [1,2]
first_second_class = titanic[titanic['Pclass'].isin([1,2])]
first_second_class.head()
```

```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
# 1. Add Fare per age
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']
```

```python
# 2. Subset > 5
high_fare_age = titanic[titanic['fare_per_year'] > 5]
```

```python
# 3. Sort descending
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
```

```python
# 4. Select name + fare per year
result = high_fare_age_srt[['Name','fare_per_year']]
result.head(10)
```

```python
# 1. Add fare per class
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']
```

```python
# 2. Adult males
adult_males = titanic[(titanic['Age'] >= 18) & (titanic['Sex'] == 'male')]
```

```python
# 3. Sort descending
adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)
```

```python
# 4. Select name + age + fare_per_class
result2 = adult_males_srt[['Name','Age','fare_per_class']]
result2.head(10)
```

```python
# 1. Total Fare
total_fare = titanic['Fare'].sum()

# 2,3. Fare by classes
fare_1 = titanic[titanic['Pclass'] == 1]['Fare'].sum()
fare_2 = titanic[titanic['Pclass'] == 2]['Fare'].sum()
fare_3 = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# 4. Combine
fare_list = [fare_1, fare_2, fare_3]

# 5. Percentage
fare_percent = [(f/total_fare)*100 for f in fare_list]
fare_percent
```

```python
# 1. Age groups
def age_group(x):
    if x < 18:
        return 'child'
    elif x < 65:
        return 'adult'
    else:
        return 'senior'

titanic['age_group'] = titanic['Age'].apply(age_group)
```

```python
# 2. Total passengers
total = len(titanic)
```

```python
# 3. Count per group
age_counts = titanic['age_group'].value_counts()
```

```python
# 4. Percent
age_percent = (age_counts / total) * 100
age_percent
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x=age_percent.index, y=age_percent.values)
plt.title("Passenger Percentage by Age Category")
plt.ylabel("Percentage (%)")
plt.xlabel("Age Group")
plt.show()
```
