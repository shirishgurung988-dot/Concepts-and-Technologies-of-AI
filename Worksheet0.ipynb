```python
#Section 7.1
#Task1
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
# Create empty lists for classifications
low_study = []
moderate_study = []
high_study = []

# Iterate over time_data and classify study hours
for day in time_data:
    study_hours = day[0]

    if study_hours < 3:
        low_study.append(study_hours)
    elif 3 <= study_hours <= 5:
        moderate_study.append(study_hours)
    else:
        high_study.append(study_hours)

# Print the classifications
print(f"Low Study Time (< 3 hours): {low_study}")
print(f"Moderate Study Time (3-5 hours): {moderate_study}")
print(f"High Study Time (> 5 hours): {high_study}")
```

```python
# 8.1.1 Exercise- Recursion
# Task 1
def sum_nested_list(nested_list):

    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)  # Recursive call
        else:
            total += element
    return total

# Test the function
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print(f"Nested list: {nested_list}")
print(f"Sum of all elements: {result}")
```

```python
#Task 2
def generate_permutations(s):

    # Base case: if string is empty or single character
    if len(s) <= 1:
        return [s]

    # List to store all permutations
    perms = []

    # Use a set to track processed characters (for duplicates)
    processed = set()

    for i, char in enumerate(s):
        # Skip if we've already processed this character at this position
        if char in processed:
            continue
        processed.add(char)

        # Get remaining characters
        remaining = s[:i] + s[i+1:]

        # Generate permutations of remaining characters
        for perm in generate_permutations(remaining):
            perms.append(char + perm)

    return perms

# Test with "abc"
perms_abc = generate_permutations("abc")
print(f"Permutations of 'abc': {perms_abc}")

# Test with "aab"
perms_aab = generate_permutations("aab")
print(f"Permutations of 'aab': {perms_aab}")
```

```python
#Task 3

def calculate_directory_size(directory):

    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):
            # It's a subdirectory, recursively calculate its size
            total_size += calculate_directory_size(value)
        else:
            # It's a file, add its size
            total_size += value

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

total_size = calculate_directory_size(directory_structure)
print(f"Directory structure: {directory_structure}")
print(f"Total directory size: {total_size} KB")
```

```python
# SECTION 8.2.1: DYNAMIC PROGRAMMING EXERCISES
# Task 1
def min_coins(coins, amount):

    # Initialize DP array
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins needed for amount 0

    # Build up the DP table
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Coins: {coins}, Amount: {amount}")
print(f"Minimum coins needed: {result}")
```

```python
#Task 2
def longest_common_subsequence(s1, s2):
    """
    Find the length of the longest common subsequence of two strings.

    Args:
        s1 (str): First string
        s2 (str): Second string

    Returns:
        int: Length of LCS
    """
    m, n = len(s1), len(s2)

    # Create DP table
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    # Fill the DP table
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[m][n]

# Test the function
s1 = "abcde"
s2 = "ace"
lcs_length = longest_common_subsequence(s1, s2)
print(f"String 1: '{s1}'")
print(f"String 2: '{s2}'")
print(f"Length of LCS: {lcs_length}")
```

```python
#Task3
def knapsack(weights, values, capacity):

    n = len(weights)

    # Create DP table
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Fill the DP table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                # Can include this item
                dp[i][w] = max(
                    values[i-1] + dp[i-1][w - weights[i-1]],  # Include item
                    dp[i-1][w]  # Exclude item
                )
            else:
                # Cannot include this item
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]

# Test the function
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
max_value = knapsack(weights, values, capacity)
print(f"Weights: {weights}")
print(f"Values: {values}")
print(f"Capacity: {capacity}")
print(f"Maximum value achievable: {max_value}")
```
